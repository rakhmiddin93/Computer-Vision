```python
def dataset(path_list:list)->tuple:
  '''
  return points to lists
  return the images to the lists
  '''
  bbox_list = []
  img_list = []
  for i in path:
    path_list.append(i)
  for j in path_list:
    raw_text=open(j).readlines()
    for line in raw_text:
      if 'Image filename : ' in line :
        path_string=line.split(' : ')[1]
        image_path=path_string.strip('/n')[1:-2]
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_list.append(img)
      elif 'Xmin' in line:
        bbox_string=line[:-1].split(' : ')[-1].split(' - ')
        minxy=tuple(map(int,[i.replace('(','').replace(')','') for i in bbox_string[0].split(',')]))
        maxxy=tuple(map(int,[i.replace('(','').replace(')','') for i in bbox_string[1].split(',')]))
        bbox_list.append((minxy,maxxy))
  return bbox_list,img_list
```
