```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import json
```

```python
def get_points(data):
    number_list = []
    for i in data[20:36]:
        number_list.append(int(float(i.strip().strip(','))))
    number_array = np.array(number_list)
    points = np.resize(number_array, (8, 2))
    return points
get_points(data)
```

```python
def get_mask(img_shape,points):
  mask_img = np.zeros(img_shape, dtype=np.uint8)
  mask_img = cv2.fillPoly(mask_img, [points], color=(255,255,255))
  
  return mask_img
```

```python
def get_ROI(img,mask):
  ROI_img = cv2.bitwise_and(img, mask)  

  return ROI_img
```

```python
filename = 'COCO.json'
f = open(filename).read()
data = f.split('\n')
points = get_points(data)
print(type(points))
print(points.shape)
```

```python
img = cv2.imread('bus_stop.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_shape = img.shape

img_mask = get_mask(img_shape,points)
```

```python
plt.imshow(img_mask)
plt.show()
```

```python
ROI_img = get_ROI(img,img_mask)
plt.imshow(ROI_img)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.imshow(np.hstack((img, img_mask, ROI_img)))
plt.axis('off')
plt.show()
```

```python
img1 = img_challenge[60:480, 300:730, :]
img2 = img_challenge[450:870, 80:510, :]
img3 = img_challenge[450:870, 500:930, :]

plt.figure(figsize=(12,12))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(img3)
plt.show()
```
